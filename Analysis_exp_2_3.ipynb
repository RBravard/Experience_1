A first way to look for stability of the biases is to compute the correlation between the sessions. Correlation between sessions 2 and 3 can give us a view of fiability over 24 hours : correlation with session 1 gives us the stability after a month.

```python
#Uncomment the correlation type, p is pearson ans s is spearman

corr_type='p'
#corr_type='s'
```

```python
#correlation for sessions 1 and 2 (stability)

from scipy import stats

corr1=np.zeros(n)
corr_mean1=0
for k in range(n):
    if corr_type=='p':
        corr1[k]=stats.pearsonr(weights2[k],weights1[k])[0]
        corr_mean1+=corr1[k]
    elif corr_type=='s':
        corr1[k]=stats.spearmanr(weights2[k],weights1[k])[0]
        corr_mean1+=corr1[k]
corr_mean1=corr_mean1/n
    
```

```python
#correlation for sessions 2 and 3 (fiability)

corr2=np.zeros(n)
corr_mean2=0
for k in range(n):
    if corr_type=='p':
        corr2[k]=stats.pearsonr(weights2[k],weights3[k])[0]
        corr_mean2+=corr2[k]
    elif corr_type=='s':
        corr2[k]=stats.spearmanr(weights2[k],weights3[k])[0]
        corr_mean2+=corr2[k]
corr_mean2=corr_mean2/n
```
